```python
import random

# Extended character set
CHARSET = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789!@#$%^&*()-_=+[]{};:,.<>?/ "

# =======================
# Key Generation (OTP)
# =======================
def generate_key(length):
    key = ""
    for _ in range(length):
        key += random.choice(CHARSET)
    return key


# =======================
# Encryption Function
# =======================
def stringEncryption(text, key):
    cipherText = ""

    for i in range(len(text)):
        # Get index of plaintext and key characters
        p_index = CHARSET.index(text[i])
        k_index = CHARSET.index(key[i])

        # Encryption using modulo arithmetic
        c_index = (p_index + k_index) % len(CHARSET)

        # Convert index back to character
        cipherText += CHARSET[c_index]

    return cipherText


# =======================
# Decryption Function
# =======================
def stringDecryption(cipherText, key):
    plainText = ""

    for i in range(len(cipherText)):
        # Get index of cipher and key characters
        c_index = CHARSET.index(cipherText[i])
        k_index = CHARSET.index(key[i])

        # Decryption using modulo arithmetic
        p_index = (c_index - k_index) % len(CHARSET)

        # Convert index back to character
        plainText += CHARSET[p_index]

    return plainText


# =======================
# User Input Section
# =======================
plainText = input("Enter Plain Text: ")

# Validation: ensure all characters are in CHARSET
if any(ch not in CHARSET for ch in plainText):
    print("Error: Plain text contains unsupported characters.")
else:
    # Auto-generate OTP key
    key = generate_key(len(plainText))

    # Encrypt
    encryptedText = stringEncryption(plainText, key)

    # Decrypt
    decryptedText = stringDecryption(encryptedText, key)

    # Output results
    print("\nGenerated Key  :", key)
    print("Cipher Text   :", encryptedText)
    print("Decrypted Text:", decryptedText)
```
